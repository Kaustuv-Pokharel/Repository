```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

```python
df=pd.read_csv("E:\\IdeaProjects\\python\\notebook projects\\exercise\\canada_per_capita_income.csv")
df.head(3)
```

```python
ax=sns.scatterplot(x='year',y='capita' , data=df,color='blue',marker='*')
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']], df['capita'])#we should supply 2d array in the x variable
```

```python

reg.predict([[2020]])
```

```python
predicted=reg.predict(df[['year']])

sns.scatterplot(x='year',y='capita',data=df)
sns.lineplot(x=df.year,y=predicted,color='red')
plt.title("Canada's per capita yearly in usd($)")
plt.show()
```
